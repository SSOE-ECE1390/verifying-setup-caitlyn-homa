# Load a Grayscale Image

```python
import cv2
import os
import numpy as np
import matplotlib.pyplot as plt

#Read image in using openCV - had trouble using the os.path.relpath
apollo11_image = cv2.imread('/Users/caitlynhoma/github-classroom/SSOE-ECE1390/verifying-setup-caitlyn-homa/Data/Apollo_11_Launch.jpg')
```

# View the Image

```python
#Print size of image
print("Image size (H, W) is:", apollo11_image.shape)

#Print image data-type 
print("Data type of image is:", apollo11_image.dtype)

#Draw image in notebook
plt.imshow(apollo11_image)
```

# Crop the Image

```python
#Select pixel region that is being kept (only middle region with rocket)
cropped_region = apollo11_image[50:650, 400:800]

#Draw cropped image
plt.imshow(cropped_region)
```

# Edit Brightness of the Image

```python
#Create a matrix to scale the brightness of Apollo 11 image
matrix = np.ones(apollo11_image.shape, dtype="uint8") * 50

#Create a brighter image
apollo11_brighter = cv2.add(apollo11_image, matrix)

#Create a darker image 
apollo11_darker = cv2.subtract(apollo11_image, matrix)

#Show brighter, original, and darker image
plt.figure(figsize=[18, 5])
plt.subplot(131); plt.imshow(apollo11_brighter) ; plt.title("Brighter")
plt.subplot(132); plt.imshow(apollo11_image) ; plt.title("Original")
plt.subplot(133); plt.imshow(apollo11_darker); plt.title("Darker")

```

# Rotate the Image

```python
#Create image to flip horizontally around x-axis
apollo11_flipped_horz = cv2.flip(apollo11_image, 1)

#Create image to flip vertically around y-axis 
apollo11_flipped_vert = cv2.flip(apollo11_image, 0)

#Create image to flip around both x- and y- axis
apollo11_flipped_both = cv2.flip(apollo11_image, -1)

#Draw the original, horz, vert, and both images
plt.figure(figsize=(18, 5))
plt.subplot(141); plt.imshow(apollo11_image); plt.title("Original")
plt.subplot(142); plt.imshow(apollo11_flipped_horz); plt.title("Horizontal Flip")
plt.subplot(143); plt.imshow(apollo11_flipped_vert); plt.title("Vertical Flip")
plt.subplot(144); plt.imshow(apollo11_flipped_both); plt.title("Both Flipped")
```

# Save the Modified Image

```python
#Saving cropped image
cv2.imwrite("cropped_apollo11.png", cropped_region)

#Declutter github repo
os.remove("cropped_apollo11.png")
```
